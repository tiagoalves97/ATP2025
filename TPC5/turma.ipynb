```python
def criar_turma():
    turma = []
    print("Turma criada.")
    return turma


def inserir_aluno(turma):
    nome = input("Nome do aluno: ")
    id = input("ID do aluno: ")

    notas = []
    etiquetas = ["Nota TPC", "Nota Projeto", "Nota Teste"]
    i = 0
    while i < 3:
        try:
            nota = float(input(f"{etiquetas[i]}: "))
            if nota < 0 or nota > 20:
                print("A nota deve ser entre 0 e 20.")
            else:
                notas.append(nota)
                i += 1
        except ValueError:
            print("Valor inválido. Introduz um número.")

    aluno = (nome, id, notas)
    turma.append(aluno)
    print(f"✅ Aluno {nome} inserido.")


def listar_turma(turma):
    if len(turma) == 0:
        print("A turma está vazia.")
    else:
        print("\n - Lista de Alunos - ")
        for aluno in turma:
            print(f"Nome: {aluno[0]} | ID: {aluno[1]} | Notas: {aluno[2]}")
        print("-----------------------")


def consultar_aluno(turma):
    id = input("ID do aluno a consultar: ")
    encontrado = False
    for aluno in turma:
        if aluno[1] == id:
            print(f"Nome: {aluno[0]} | ID: {aluno[1]} | Notas: {aluno[2]}")
            encontrado = True
    if not encontrado:
        print("Aluno não encontrado.")


def guardar_turma(turma):
    nome_fich = input("Nome do ficheiro para guardar: ")
    try:
        with open(nome_fich, "w") as f:
            for aluno in turma:
                nome, id, notas = aluno
                linha = f"{nome};{id};{notas[0]};{notas[1]};{notas[2]}\n"
                f.write(linha)
        print("Turma guardada.")
    except Exception as e:
        print(f"Erro ao guardar a turma: {e}")


def carregar_turma():
    nome_fich = input("Nome do ficheiro a carregar: ")
    turma = []
    try:
        with open(nome_fich, "r") as f:
            for linha in f:
                dados = linha.strip().split(";")
                if len(dados) == 5:
                    nome = dados[0]
                    id = dados[1]
                    try:
                        notas = [float(dados[2]), float(dados[3]), float(dados[4])]
                        aluno = (nome, id, notas)
                        turma.append(aluno)
                    except ValueError:
                        print(f"Aviso: linha inválida ignorada: {linha.strip()}")
        print("Turma carregada.")
    except FileNotFoundError:
        print("Ficheiro não encontrado. Turma vazia criada.")
    except Exception as e:
        print(f"Erro ao carregar turma: {e}")
    return turma

turma = []
x = True
while x:
    print("\n--- Gestão de Alunos ---")
    print("1: Criar uma turma")
    print("2: Inserir um aluno na turma")
    print("3: Listar a turma")
    print("4: Consultar um aluno por id")
    print("8: Guardar a turma em ficheiro")
    print("9: Carregar uma turma dum ficheiro")
    print("0: Sair da aplicação")

    op = input("Opção: ")

    if op == "1":
        turma = criar_turma()
    elif op == "2":
        inserir_aluno(turma)
    elif op == "3":
        listar_turma(turma)
    elif op == "4":
        consultar_aluno(turma)
    elif op == "8":
        guardar_turma(turma)
    elif op == "9":
        turma = carregar_turma()
    elif op == "0":
        print("A sair da aplicação...")
        x = False
    else:
        print("Opção inválida. Tente novamente.")
```
