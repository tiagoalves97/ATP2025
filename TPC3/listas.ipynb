```python
import random

def lista_aleatoria():
    lista = []
    while True:
        try:
            n = int(input("Quantos números quer na sua lista? (1 - 100)"))
            if 0 < n < 101:
                lista = [random.randint(1, 100) for i in range(n)]
                return lista
            else:
               print("Tem de ser um número entre 1 e 100!")
        except ValueError:
            print("Tem de ser um número!")

def lista_utilizador():
    lista = []
    print("Introduz números inteiros, quando desejares terminar escreve - Fim ")
    while True:
        entrada = input("Número: ")
        if entrada == "Fim":
            break
        try:
            elem = int(entrada)
            lista.append(elem)
        except ValueError:
            print("Só é valido números inteiros!")  
    return lista           
        

def soma(lista):
    return sum(lista)

def media(lista):
    return sum(lista) / len(lista) if lista else 0

def maior(lista):
    return max(lista)

def menor(lista):
    return min(lista)

def procura(lista):
    try:
        alvo = int(input("Número a procurar: "))
        posicoes = []
        indice = 0
        for valor in lista:
            if valor == alvo:
                posicoes.append(indice + 1)
            indice += 1
        if posicoes:
            print(f"O número {alvo} está na {posicoes} posição.")
        else:
            print("-1 (não encontrado)")
    except ValueError:
        print("Tem de usar um número inteiro.")


print(""" 
          (1) Criar lista
          (2) Ler lista
          (3) Soma
          (4) Média
          (5) Maior
          (6) Menor
          (7) Ordenada por ordem cresente
          (8) Ordenada por ordem decresente
          (9) Procura um elemento
          (0) Sair""")

while True:
    escolha = input("Escolha uma ação")
    if escolha == "1":
        lista = lista_aleatoria()
        print(lista)
    elif escolha == "2":
        lista = lista_utilizador()
        print(lista)
    elif escolha == "3":
        if lista:
            print(soma(lista))
        else:
            print("Não há lista.")
    elif escolha == "4":
        if lista:
            print(media(lista))
        else:
            print("Não há lista.")
    elif escolha == "5":
        if lista:
            print(max(lista))
        else:
            print("Não há lista.")
    elif escolha == "6":
        if lista:
            print(min(lista))
        else:
            print("Não há lista.")
    elif escolha == "7":
        if lista:
            print(sorted(lista))
        else:
            print("Não há lista.")
    elif escolha == "8":
        if lista:
            print(sorted(lista)[::-1])
        else:
            print("Não há lista.")
    elif escolha == "9":
        if lista:
            procura(lista)
        else:
            print("Não há lista.")
    elif escolha == "0":
        print("Muito bem! Volta quando quiseres!")
        break
    else:
        print("Escolhe uma das opções disponíveis!")

```
